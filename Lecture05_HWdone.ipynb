**1. Нормализовать данные**

```python
from sklearn.preprocessing import StandardScaler, PolynomialFeatures
from sklearn.linear_model import LogisticRegression, LogisticRegressionCV
from sklearn.pipeline import Pipeline
from sklearn.model_selection import train_test_split
from sklearn.metrics import roc_auc_score

#признаки
features = df.loc[:, "MFCCs_ 1":"MFCCs_22"]
families = df['Family'].unique()

#нормализуем
scaler = StandardScaler()
normalized_features = scaler.fit_transform(features)
```

**2. Построить бинарный классификатор для каждого семейства(Family) на основе LogisticRegression c $l_1$ и $l_2$ регуляризацими. Это соответствует целевым исходам (Конкретное семейство, Всё остальное). Подобрать гиперпараметр $C$ с помощью кросс-валидации, используя метрику ROC AUC.**

```python
#строим классификаторы для каждого семейства
results = {}
for family in families:
    print(f"\n--- Обработка семейства: {family} ---")

    y = (df['Family'] == family).astype(int)

    #разделим
    X_train, X_test, y_train, y_test = train_test_split(
        normalized_features, y, test_size=0.3, random_state=42
    )

    for penalty in ['l1', 'l2']:
        solver = 'liblinear' if penalty == 'l1' else 'lbfgs'


        model = LogisticRegressionCV(
            Cs=10,
            cv=5,
            scoring='roc_auc',
            penalty=penalty,
            solver=solver,
            max_iter=10000,
            random_state=42
        )
        model.fit(X_train, y_train)

        best_C = model.C_[0]
        y_pred = model.predict_proba(X_test)[:, 1]
        auc = roc_auc_score(y_test, y_pred)

        key = f"{family} ({penalty.upper()})"
        results[key] = {'C': best_C, 'AUC': auc}
        print(f"{key}: C = {best_C:.4f}, AUC = {auc:.4f}")
```

**3. Добавить полиномиальные признаки до второго порядка. Построить бинарный классификатор для каждого семейства на основе LogisticRegression c $l_1$ и $l_2$ регуляризацими. Подобрать гиперпараметр $C$ с помощью кросс-валидации, используя метрику ROC AUC.**

```python
#добавим полиномиальные признаки
poly = PolynomialFeatures(degree=2, include_bias=False)
poly_features = poly.fit_transform(normalized_features)

for family in families:
    print(f"\n--- Обработка семейства с полиномиальными признаками: {family} ---")

    #бинарная целевая переменная
    y = (df['Family'] == family).astype(int)

    #разделение данных
    X_train, X_test, y_train, y_test = train_test_split(
        poly_features, y, test_size=0.3, random_state=42
    )

    for penalty in ['l1', 'l2']:
        solver = 'liblinear' if penalty == 'l1' else 'saga'

        model = Pipeline([
            ('scaler', StandardScaler()),
            ('clf', LogisticRegressionCV(
                Cs=10,
                cv=5,
                scoring='roc_auc',
                penalty=penalty,
                solver=solver,
                max_iter=5000,
                random_state=42
            ))
        ])
        model.fit(X_train, y_train)


        best_C = model.named_steps['clf'].C_[0]
        y_pred = model.predict_proba(X_test)[:, 1]
        auc = roc_auc_score(y_test, y_pred)


        key = f"{family} ({penalty.upper()}) + Poly"
        results[key] = {'C': best_C, 'AUC': auc}
        print(f"{key}: C = {best_C:.4f}, AUC = {auc:.4f}")

#выводим итог
print("\n--- Итоговые результаты ---")
for key, res in results.items():
    print(f"{key}: Оптимальное C = {res['C']:.4f}, ROC AUC = {res['AUC']:.4f}")
```
